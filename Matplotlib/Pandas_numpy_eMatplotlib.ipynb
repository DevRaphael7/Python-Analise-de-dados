```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tabela = pd.read_excel('Tabela\Preço por estado.xlsx')
tabela
```

```python
# Adicionando as colunas ao gráfico matplotlib
plt.plot(tabela.Produtos, tabela['Acre'], 'b.--')
plt.plot(tabela.Produtos, tabela['Paraná'], 'g--')
plt.plot(tabela.Produtos, tabela['Bahia'], '.-')
plt.plot(tabela.Produtos, tabela['São Paulo'], 'g.-')
plt.plot(tabela.Produtos, tabela['Rio de Janeiro'], 'r.-')

#Criando legendas para meu gráfico matplotlib
plt.legend(['Acre', 'Paraná', 'Bahia', 'São Paulo', 'Rio de Janeiro'])
plt.title('Preço de produtos por estado\n')

#Obter somente as colunas de minha tabela Excel
print(tabela.columns)

#Um título ou descrição para os eixos x e y
plt.xlabel('\nProdutos', fontsize=12)
plt.ylabel('Preço\n', fontsize=12)
plt.show()
```

```python
# Usando estrutura de repetição para visualizar o gráfico

plt.title("Preço de produtos por estado\n", fontweight='bold', fontsize=20)

lista = []
lista_estado = ['Acre', 'Paraná', 'Bahia', 'São Paulo', 'Rio de Janeiro']

for estado in lista_estado:
    plt.plot(tabela['Produtos'], tabela[estado], '.-')
    acumulador = estado
    lista.append(acumulador)
    plt.legend(lista)
    
'''
    Em casos do número de colunas for muito grande, é melhor usar uma estrutura de repetição, será mais 
    fácil.

'''
plt.yticks([0, 5.0, 10.0, 15.0, 20.0, 25.0, 30.0, 35.0, 40.0])
plt.savefig('preco_de_produtos_por_estado.png', dpi=300)
plt.show()
```

```python
tabela_dois = pd.read_excel('Tabela/notas semestre.xlsx')

#Com head( ) posso definir a quantidade de linhas que quero ver da minha tabela
tabela_dois.head()
```

### Histograma

```python
bins = [200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420, 440, 460, 480, 500]
plt.hist([tabela_dois['1° semestre']], bins=bins)

plt.xlabel('\nNotas do primeiro semestre', fontsize=20)
plt.show()
```

### Gráfico de pizza

```python
maior_trezentos = tabela_dois.loc[tabela_dois['2 ° Semestre'] >= 350].count()[0]
menor_trezentos = tabela_dois.loc[tabela_dois['2 ° Semestre'] < 350].count()[0]

plt.pie([maior_trezentos, menor_trezentos], colors=['#5A2CEE', '#A42070'], autopct='%.2f %%')
plt.title('Desempenho das salas no 2° semestre')
plt.legend(['Acima dos trezentos e cinquenta', 'Abaixo dos trezentos e cinquenta'])
plt.show()
```
